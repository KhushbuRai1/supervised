```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
```

```python
#Import and store the data in a data frame. 
url = 'http://users.stat.ufl.edu/~winner/data/airq402.dat'
df = pd.DataFrame(pd.read_csv(url, header = None, sep='\s+')).apply(pd.to_numeric, errors = 'ignore')
```

```python
#2.Remove the outliers from the data 
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
mincap = (Q1-1.5*IQR)
maxcap = (Q3+1.5*IQR)
df1.apply(pd.to_numeric, errors = 'ignore')
df1 = df1[(df1 >= mincap) & (df1 <= maxcap)]
   # print(df.loc[(df[column]>=mincap) & (df[column]<=maxcap),:])
df1 = df1.dropna(axis = 0)
df1.shape
df1
```

```python
#6.Divide the data set into training and test data set and build a Multiple Linear Regression model. 
X = df1.drop([2], axis = 1)
y = df1[2]
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(X_train,y_train)
```

```python
#Print the coefficients & intercepts of the linear regression model 
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[idx]))
```

```python
regression_model.score(X_test, y_test)
    
```

```python
regression_model.score(X_train, y_train)
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```
